```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import numpy.linalg as ln
import cv2

from matplotlib import rcParams
rcParams['figure.figsize'] = (12, 6)
rcParams['figure.dpi'] = 150
rcParams['font.size'] = 16
rcParams['font.family'] = 'sans-serif'
rcParams['axes.facecolor'] = '#ffffff'
rcParams['lines.linewidth'] = 2.0
```

```python
# Using three test images\n",
# im1 = lit right\n",
# im2 = lit left above\n",
# im3 = lit left below\n",
# height = 130.0
height = 100.0
L1 = np.array([[-22],[0],[height]],np.float)
L1 = L1/ln.norm(L1)
L2 = np.array([[15],[-18],[height]],np.float)
L2 = L2/ln.norm(L2)
L3 = np.array([[15],[18],[height]],np.float)
L3 = L3/ln.norm(L3)

L = np.hstack((L1, L2, L3))
Linv = ln.inv(L)
print Linv
```

```python
im1 = cv2.imread('./res/coins/im3-1.png')
im2 = cv2.imread('./res/coins/im1-1.png')
im3 = cv2.imread('./res/coins/im2-1.png')
plt.imshow(im1)
plt.show()
plt.imshow(im2)
plt.show()
plt.imshow(im3)
plt.show()
```

```python
channel = 0;
imB = np.dstack((im1[:,:,channel],im2[:,:,channel],im3[:,:,channel]))
imBH = imB.copy()

# Set crop area
x1 = 100
x2 = 900
y1 = 100
y2 = 1500

# Set up output arrays
xGrad = np.zeros(imB[:,:,0].shape, np.float)
yGrad = np.zeros(imB[:,:,0].shape, np.float)
s     = np.zeros(imB[:,:,0].shape, np.float)

# Blur images for high pass filtering
blurSig = 100
imBlur1 = cv2.GaussianBlur(imB[:,:,0],None,blurSig)
imBlur2 = cv2.GaussianBlur(imB[:,:,1],None,blurSig)
imBlur3 = cv2.GaussianBlur(imB[:,:,2],None,blurSig)

imBH[:,:,0] = imB[:,:,0] - imBlur1
imBH[:,:,1] = imB[:,:,1] - imBlur2
imBH[:,:,2] = imB[:,:,2] - imBlur3

# Take a crop of the array
imBCrop = imBH[x1:x2,y1:y2,:]

# For each pixel calculate the three outputs
for iX, imLine in enumerate(imBCrop):
    for iY, imPix in enumerate(imLine):
        
        I = np.array(np.reshape(imPix,(3,1)),np.float)
        M = np.dot(Linv,I)
        
        xGrad[iX + x1, iY + y1] = M[0]/M[2]
        yGrad[iX + x1, iY + y1] = M[1]/M[2]
        s[iX + x1, iY + y1]     = ln.norm(M)

# plot the outputs
plt.imshow(np.clip(xGrad[x1:x2,y1:y2],-1,1))
plt.colorbar()
plt.show()
plt.imshow(np.clip(yGrad[x1:x2,y1:y2],-1,1))
plt.colorbar()
plt.show()
plt.imshow(np.clip(s[x1:x2,y1:y2],0,5000))
plt.colorbar()
plt.show()
```

```python
cv2.imwrite('./res/dx.png',127*xGrad + 127)
cv2.imwrite('./res/dy.png',127*yGrad + 127)
```
