```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import numpy.linalg as ln
import cv2

from matplotlib import rcParams
rcParams['figure.figsize'] = (12, 6)
rcParams['figure.dpi'] = 150
rcParams['font.size'] = 16
rcParams['font.family'] = 'sans-serif'
rcParams['axes.facecolor'] = '#ffffff'
rcParams['lines.linewidth'] = 2.0
```

```python
# Using three test images\n",
# im1 = lit approx from camera\n",
# im2 = lit from above\n",
# im3 = lit from left\n",
L1 = np.array([[-22],[0],[130]])
L1 = L1/ln.norm(L1)
L2 = np.array([[15],[-18],[130]])
L2 = L2/ln.norm(L2)
L3 = np.array([[15],[18],[130]])
L3 = L3/ln.norm(L3)

L = np.hstack((L1, L2, L3))
Linv = ln.inv(L)
```

```python
im1 = cv2.imread('./res/im1.png')
im2 = cv2.imread('./res/im2.png')
im3 = cv2.imread('./res/im3.png')
plt.imshow(im1)
plt.show()
plt.imshow(im2)
plt.show()
plt.imshow(im3)
plt.show()
```

```python
imB = np.dstack((im1[:,:,0],im2[:,:,0],im3[:,:,0]))
I = np.array(np.reshape(imB[100,100,:],(3,1)),np.float)
M = np.dot(Linv,I)
print M[0]/M[2]
print M[1]/M[2]
```

```python
x1 = 50
x2 = 500
y1 = 50
y2 = 1000
xGrad = np.zeros(imB[:,:,0].shape)
yGrad = np.zeros(imB[:,:,0].shape)
s = np.zeros(imB[:,:,0].shape)
imBCrop = imB[x1:x2,y1:y2,:]
for iX, imLine in enumerate(imBCrop):
    for iY, imPix in enumerate(imLine):
        I = np.array(np.reshape(imPix,(3,1)),np.float)
        M = np.dot(Linv,I)
        xGrad[iX + x1, iY + y1] = M[0]/M[2]
        yGrad[iX + x1, iY + y1] = M[1]/M[2]
        s[iX + x1, iY + y1] = ln.norm(M)
```

```python
plt.imshow(np.clip(xGrad[x1:x2,y1:y2],-10,10))
plt.colorbar()
plt.show()
plt.imshow(np.clip(yGrad[x1:x2,y1:y2],-10,10))
plt.colorbar()
plt.show()
plt.imshow(s[x1:x2,y1:y2])
plt.colorbar()
plt.show()
```

```python
cv2.imwrite('./res/dx.png',127*xGrad + 127)
cv2.imwrite('./res/dy.png',127*yGrad + 127)
```
