```python
%matplotlib inline
import numpy as np
import cv2
import matplotlib.pyplot as plt
from scipy.cluster.vq import kmeans
```

# Getting started
Lets try loading an image with OpenCV and displaying with inline with matplotlib

```python
im = cv2.imread('res/giant.jpg')
im = cv2.resize(im, None, fx=0.25, fy=0.25)
plt.imshow(cv2.cvtColor(im,cv2.COLOR_BGR2RGB))
```

```python
means = np.empty((1,3))
cols = np.reshape(im,(-1,3))
res,d = kmeans(cols, 10)
means = np.row_stack((means,res))
means = np.array(means)
```

## Next
Now we have the clustered means, lets plot them to see if they make sense

```python
temp = np.zeros((100,400,3),np.uint8)
temp[:,:99,:] = means[5,:]
temp[:,100:199,:] = means[6,:]
temp[:,200:299,:] = means[2,:]
temp[:,300:,:] = means[7,:]
plt.imshow(cv2.cvtColor(temp,cv2.COLOR_BGR2RGB))
```

```python
f = plt.figure(figsize=(8,8), dpi=100)
axes = f.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height (range 0 to 1)
axes.plot(im[:,:,0],im[:,:,1],'.')
axes.set_aspect('equal')
```
